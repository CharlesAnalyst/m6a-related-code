```python
def morethan_cutoff(m6a_id, tissue, term): # term: adult or fetus
    cols = df_matrix.columns
    selected_cols = [x for x in cols if (tissue in x) and (term in x)]
    values = df_matrix.loc[m6a_id]
    ratio_list = [values[x] for x in selected_cols]
    bool_v = False
    if np.mean(ratio_list) > 1.5:
        bool_v = True
    return bool_v
```
