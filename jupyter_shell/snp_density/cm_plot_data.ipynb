```python
def process_single_file(x_file):
    with open(i_file, 'r') as f:
        seq = f.read().strip()
        pos_list = [i for i in range(0, len(seq)-1, 1) if seq[i:i+2] == "cg"]
        cg_name_list = [("cg_%d" % i) for i in range(len(pos_list))]
        chr_name_list = [chr_name for i in range(len(pos_list))]
        new_pos_list = chunks(pos_list, seq, len(seq) / BIN_NUMBER)
        print(len(pos_list), len(cg_name_list), len(chr_name_list), len(new_pos_list))
        df = pd.DataFrame({"CG": cg_name_list, "chr": chr_name_list, "position": new_pos_list})
    return df
```

```python
def chunks(position_list, seq, n):
    print(n)
    count, new_position_list = 0, []
    for i in np.arange(0, len(seq), n):
        left_border, right_border = i, i + n
#         print(left_border, right_border)
        sub_list = [j for j in position_list if left_border <= j < right_border]
        ratio_list = [(x-left_border)/(right_border - left_border) for x in sub_list]
        simulated_left, simulated_right = INTERVAL*count, INTERVAL*(count+1)
        simulated_list = [int(simulated_left + INTERVAL*x) for x in ratio_list]
        count += 1
#         simulated_list = []
#         for x in sub_list:
#             r = random.randint(simulated_left, simulated_right)
#             if r not in simulated_list: simulated_list.append(r)
        new_position_list += simulated_list
        for x in range(len(sub_list)):
            print(sub_list[x], simulated_list[x])
#     print(len(position_list), len(new_position_list))
#     print(position_list)
#     print(new_position_list)
    return new_position_list
```
