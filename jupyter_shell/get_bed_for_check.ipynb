```python
import os
import sys
import shutil
import glob
import pandas as pd
import numpy as np
from scipy import stats
import subprocess
import base64
```

```python
file_dir = "/data5/galaxy/project/expre_vs_m6a"
gene_bed = "/data/database/GRCh38/GENCODE/Genes_ensembl.bed"
df_gene = pd.read_table(gene_bed, sep="\t", header=None, names=["a", "b", "d", "name", "e", "f"])
m6a_dir = "/data5/galaxy/project/methyl_m6a/data/diff_m6a_peak/macs2_bdgdiff/combination/"
result_dir = "/data5/galaxy/project/expre_vs_m6a/result_bed"

file_list = glob.glob("%s/*.txt" % file_dir)
for i_file in file_list:
    print(os.path.basename(i_file))
    df_name = pd.read_table(i_file, header=None, names=["name"])
    sub_dir = os.path.basename(i_file).split("-")[0]
    m6a_type = os.path.basename(i_file).split("_")[-1].split(".txt")[0]
    df = df_name.merge(df_gene, on="name", how="left").dropna()[["a", "b", "d", "name", "e", "f"]].sort_values(["a", "b"])
    tmp_file = "%s/%s" % (result_dir, os.path.basename(i_file).replace(".txt", "_tmp.bed"))
    df.to_csv(tmp_file, sep="\t", header=None, index=False)
#     df_str = df.to_string(header=False, index=False, col_space=4)
#     df_str = "\n".join(["\t".join(x.split()) for x in df_str.split("\n")])
#     df_str = base64.encodebytes(df_str.encode())
    result_file = "%s/%s" % (result_dir, os.path.basename(i_file).replace(".txt", ".bed"))
    if m6a_type == "up":
        m6a_file = "%s/%s/diff_peak_c3.0_cond1.bed" % (m6a_dir, sub_dir)
        get_result(m6a_file, result_file, df_str, tmp_file)
    elif m6a_type == "common":
        m6a_file = "%s/%s/diff_peak_c3.0_common.bed" % (m6a_dir, sub_dir)
        get_result(m6a_file, result_file, df_str, tmp_file)
    else:
        m6a_file = "%s/%s/diff_peak_c3.0_cond2.bed" % (m6a_dir, sub_dir)
        get_result(m6a_file, result_file, df_str, tmp_file)
```

```python
def get_result(m6a_file, result_file, df_str, tmp_file):
#     command = "bedtools intersect -a stdin -b %s -wa -wb > %s" % (m6a_file, result_file)
#     print(df_str)
#     command = "bedtools intersect -a stdin -b %s -wa -wb > %s" % (m6a_file, result_file)
#     sub_pro = subprocess.Popen(command, shell=True, stdin=subprocess.PIPE)
#     sub_pro.communicate(df_str)
    os.system("bedtools intersect -a %s -b %s -wa -wb > %s" % (tmp_file, m6a_file, result_file))
    os.remove(tmp_file)
```
